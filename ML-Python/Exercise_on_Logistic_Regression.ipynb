```python
import seaborn as sns
```

```python
sns.boxplot(df[' Plasma glucose concentration'])
```

```python
sns.boxplot(df[' Body mass index'])
```

```python
q1=df[' Body mass index'].quantile(0.25)
```

```python
q3=df[' Body mass index'].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
lower = q1 - 1.5*iqr
higher = q3 + 1.5 * iqr
```

```python
med = df[' Body mass index'].median()
```

```python
for i in df[' Body mass index']:
    if i < lower or i > higher:
        df[' Body mass index'] = df[' Body mass index'].replace(i,med)
        
```

#5.Model Training

#5.1 - > Split the dataset into training and testing sets.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

5.2 Using Logistic Regression to train the model on the training data.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state=42)
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

5.3 Predict the training model on test data

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score = accuracy_score(y_test, model.predict(X_test))
```

```python
print("Accuracy:", accuracy_score)
```
